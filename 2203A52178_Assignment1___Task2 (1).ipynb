```python
import matplotlib.pyplot as plt

#2. Generate Health Care Dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 3. Scatter Plot
plt.scatter(age, height, color='blue', label='Height')
plt.scatter(age, weight, color='red', label='Weight')
plt.title('Scatter Plot of Age vs. Height and Weight')
plt.xlabel('Age')
plt.ylabel('Measurement')
plt.legend()
plt.show()

# 4. Bar Chart
fig, ax = plt.subplots()
bar_width = 0.35
index = range(len(people))

bar1 = ax.bar(index, height, bar_width, label='Height')
bar2 = ax.bar([i + bar_width for i in index], weight, bar_width, label='Weight')

ax.set_title('Bar Chart of Health Measurements')
ax.set_xlabel('People')
ax.set_ylabel('Measurement')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(people)
ax.legend()
plt.show()

# 5. Histogram
plt.hist(age, bins=5, color='green', edgecolor='black')
plt.title('Histogram of Ages in Health Care Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
